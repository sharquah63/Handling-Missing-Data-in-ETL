SECTION A – THEORETICAL QUESTIONS

Q1. What are the most common reasons for missing data in ETL pipelines?

-> Missing data in ETL pipelines often arises from data entry errors, system failures during extraction, incomplete source records, or integration issues across heterogeneous databases.

Q2. Why is blindly deleting rows with missing values considered a bad practice in ETL?

-> Blind deletion reduces dataset size, potentially introducing bias if missingness correlates with key patterns, and loses valuable information that could inform analysis.

Q4. Why is median imputation preferred over mean imputation for skewed data such as income?

-> Median imputation is preferred over mean for skewed data like income because the median represents the central value unaffected by extreme outliers (e.g., billionaires skewing average salary high). The mean pulls toward these tails, distorting the imputed values and biasing downstream analytics.

For instance, in a dataset with incomes [10k, 20k, 30k, 1M], mean is ~265k (outlier-driven), but median 25k better reflects typical values.

Q5. What is forward fill and in what type of dataset is it most useful?

-> Forward fill (also called forward-fill imputation) replaces each missing value with the last observed non-missing value from previous records.
​

It's most useful in time-series datasets or sequentially ordered data (like sales logs, stock prices, or sensor readings) where values tend to persist or change gradually over time, preserving trends without introducing artificial jumps.

Q6. Why should flagging missing values be done before imputation in an ETL workflow?

-> Flagging missing values before imputation preserves the original missingness pattern as metadata, allowing analysts to later study why data was missing (e.g., systematic refusal patterns) without the imputation process overwriting that signal.
​

This creates an audit trail in ETL workflows: first add a binary flag column (0=present, 1=missing), then impute. It enables stratified analysis, bias detection, and regulatory compliance by distinguishing "real" vs "filled" data.

Q7. Consider a scenario where income is missing for many customers.

How can this missingness itself provide business insights?

-> High missing income for customers often signals privacy reluctance, low-income segments avoiding disclosure, or survey fatigue among high-value customers who skip sensitive questions.
​

Business insights: Segment these "non-disclosers" separately—they may represent budget-conscious buyers needing promotional offers, or premium clients valuing privacy (target with anonymized surveys). Track if missingness correlates with churn risk or low sales volume for proactive retention.
